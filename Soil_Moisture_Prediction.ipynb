# Data Integration

```python
import pandas as pd

# Reading the CSV files into dataframes
df1 = pd.read_csv('2017_final_data.csv')
df2 = pd.read_csv('2020_final_data.csv')
df3 = pd.read_csv('2023_final_data.csv')

# Concatenating the dataframes vertically
df = pd.concat([df1, df2, df3], ignore_index=True)

# Saving the merged dataframe to a new CSV file
df.to_csv('final_data.csv', index=False)

print("Merged data saved to 'final_data.csv'")
```
